```python
dimensions = 2
references = 2
datums = 3
```

```python
ref = ep.dimspace(0,1,references,dimensions)
ref.shape, ref 
```

```python
data = np.arange(datums*dimensions).reshape((datums,dimensions))
data.shape, data
```

```python
bb = np.broadcast_to(ref, (datums,references**dimensions,dimensions))
bb.shape, bb
```

```python
delt = np.moveaxis(bb, 0,1) - data
delt.shape, delt
```

```python
# (references, datums)
app = np.apply_along_axis(np.linalg.norm, 2, delt)
app.shape, app
```

```python
# then get one per reference, which are the densities
ms = np.mean(app, axis=1)
references**dimensions==ms.shape[0], ms.shape, ms
```
