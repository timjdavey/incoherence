```python
import nbsetup
import numpy as np
import pandas as pd
import ensemblepy as ep
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def distributions(count):
    return [np.random.uniform(0,1,count),
            np.random.normal(0.5, 0.05,count),
            np.random.exponential(0.1, size=count),
            np.random.normal(0.5, 0.0001, count),
            np.random.normal(0.2, 0.0001, count),
            np.ones(count)*0.2,
            ]

def plot(samples=10000, steps=1000):
    dists = distributions(samples)
    fig, axes = plt.subplots(len(dists), 2, sharex=True, sharey=False, figsize=(15,len(dists)*5))
    
    for i, d in enumerate(dists):
        sns.histplot(d, bins=10, ax=axes[i][0], stat='probability')
    
    for i, d in enumerate(dists):
        y, x = ep.densities(d, steps=steps)
        sns.lineplot(x=x.T[0], y=y, ax=axes[i][1])
```

```python
# On the left showing the discrete version of the data
# On the right showing the density approximation
# Data constrained to [0,1]
# But variance is taken for [-0.5, +1.5] to allow for drop off
plot()
```
