```python
import nbsetup
from simulations.automata.models import CA1DEnsemble
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
np.random.seed(4020)
```

```python
FILENAME = 'figures/4.1-ca-%s.eps'

def create(i, j=None, save=True):
    c = CA1DEnsemble(rule=i, cells=100, count=10, init='sparse', k=100).show(100)
    c.plot(j=j).savefig(FILENAME % i)
    return c
```

```python
c = create(218, 1)
```

```python
INCLUDE = [0, 30, 110, 220]

cas = {'218': c}
for k in INCLUDE:
    cas[k] = create(k)
```

```python
classes = {
    1: [0, ], # rapidly converge to a uniform state (zero incoherence)
    2: [4, 37, 56, 73, 108, 250], # repetitive or stable state aka lines (zero over time, but high ergodic)
    3: [22, 30, 126, 150, 182], # remain in a random state (high incoherence over time, but low ergodic)
    4: [11, 60, 110, 184, 220], # some or both
}
```

```python
cas = {}
for c, rules in classes.items():
    for i in rules:
        cas[i] = create(i)
```

```python
cas[220].plot(0,8).savefig(FILENAME % 220)
```

```python
cas[218].plot(0,6).savefig(FILENAME % 218)
```

```python
cas[184].plot(0,6, cas[184].analysis['Stable diag LR'].to_list()[-1])
```

```python

def plot_all(self, columns=5):
    import matplotlib.pyplot as plt
    rows = self.count // columns
    fig, axes = plt.subplots(columns, rows,
            sharex=False, sharey=False, figsize=(3*columns,3*rows))
    plt.tight_layout()
    aflat = axes.flatten()
    for i in range(self.count):
        self.plot_single(i, aflat[i])
    
plot_all(cas[218])
```
